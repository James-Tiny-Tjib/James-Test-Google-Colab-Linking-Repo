```python
# Code for Multiple Inputs:
# Import Libraries
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Initialize the tokenizer and model
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

# Input
sequence = ["Tokenization is extremely entertaining.","Just Kidding, but it is very important.", "River Hill is the best high school."]
# Get Input Ids
input_ids = tokenizer(sequence, return_tensors= "pt",padding = True, truncation = True)
print(input_ids)
# Feed it into model
outputs = model(**input_ids)
# Get model output
print(outputs.logits)
```

```python
# Uncovered Topic 2 - Padding and Attention Masks
# Hugging Face does a good job explaining it I'll just give a summary

# Import Necessary Libraries
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Initialize the tokenizer and model
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

### BEHIND THE SCENES OF A PADDING AND ATTENTION MASKS IN A TOKENIZER ###

# We are given the following sequence, but notice that both inputs aren't even
batched_ids = [
    [200, 200, 200],
    [200, 200]
]
# To fix this, we can put a padding token to make them the same length
padding_id = 100
batched_ids = [
    [200, 200, 200],
    [200, 200, padding_id],
]

# Because the padding, the input is rectangular and was passed through the model
# We expect the sequence1_ids and the first sequence in batched_ids to result the same (and it does)
# However even with the padding, sequence2_ids and the seocnd sequence in batched_ids do not evaluate the same
# This is because the model doesn't recognizer that the padding token is a padding token
sequence1_ids = [[200, 200, 200]]
sequence2_ids = [[200, 200]]
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]
print("Without attention mask:")
print("\tsequence1_ids: "+str(model(torch.tensor(sequence1_ids)).logits))
print("\tsequence2_ids: "+str(model(torch.tensor(sequence2_ids)).logits))
print("\tbatched_ids: "+str(model(torch.tensor(batched_ids)).logits)+"\n")

# To solve this, we can use an attention mask to tell the model know whether the token is padding (0) or not (1)
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]
# 1's represent tokens that are part of the input, while 0's represent the padding tokens
attention_mask = [
    [1, 1, 1],
    [1, 1, 0],
]
# Passing in the batch_ids and its respected attention mask
outputs = model(torch.tensor(batched_ids), attention_mask=torch.tensor(attention_mask))
# Notice now that both inputs are the same
print("With attention mask:")
print("\tsequence1_ids: "+str(model(torch.tensor(sequence1_ids)).logits))
print("\tsequence2_ids: "+str(model(torch.tensor(sequence2_ids)).logits))
print("\tbatched_ids: "+str(outputs.logits))
```
