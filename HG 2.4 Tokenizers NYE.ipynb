```python
# Word Tokenization
# Bad because the Model's dictionary would be too Long ("Dog" vs "Dogs" would require 2 tokens )
input = "Tokenization is extremely entertaining ."
input = input.split()
print(input)
```

```python
# Character Tokenization
# Bad because no context can be learned with individual characters
input = "Tokenization is extremely entertaining ."
input = [*input] # Unpacking a List
print(input)
```

```python
# Subowrd Tokenization
# Good because context is preserved, and
# Import AutoTokenizer
from transformers import AutoTokenizer
#Intialize the tokenizer and model
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
# Sequence Definition
sequence = "Tokenization is extremely entertaining ."
# Use the .tokenize() function to get the tokens
tokens = tokenizer.tokenize(sequence)
print(tokens)
```
