```python
from transformers import pipeline # From [from transformers]


# Initialize NER pipeline
ner_pipeline = pipeline("ner", model = "dslim/bert-base-NER") # From [ner_pipeline = ("NER", 'dslim/bert-base-NER")]

# Input text
text = "Stop playing Brawl Stars during class!"

# Analyze named entities
entities = ner_pipeline(text) # From [pipeline(text)]

print(entities)

output = ""
for i in range(len(entities)):
    token = entities[i]
    output +=token["word"].replace("#","")
    if((i ==(len(entities)-1)) or token["end"]!=entities[i+1]["start"]):
        output +=" "
print(output)
```
